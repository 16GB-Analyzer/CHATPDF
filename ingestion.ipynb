```python
for doc in documents:
    text = doc.page_content.lower()

    if "cnn" in text:
        doc.metadata["topic"] = "CNN"
    elif "rnn" in text:
        doc.metadata["topic"] = "RNN"
    elif "gan" in text:
        doc.metadata["topic"] = "GAN"
    else:
        doc.metadata["topic"] = "general"
```

```python
#split the documents into smaller chunks
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000,chunk_overlap=200)
texts = text_splitter.split_documents(documents)

print(f"Number of chunks created: {len(texts)}")
print(f"Content of the first chunk: {texts[0].page_content[:500]}")
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)
```

Output having negative just means that the word vector is in the opposite direction, so we just look to see similarity between vectors just matching if each vector in the vector space is almost close or similar to the other ones then they both might belong or mean somethign together

```python
#check embeddings for the first chunk
print(embeddings.embed_query(texts[0].page_content))
```

We are using semantic search to find the most relevant chunks for a given query. We will use FAISS for this purpose.

```python
#build faiss to store the embeddings
#creating FAISS index means : 
#we are converting the text chunks into vector representations and storing them in a way that allows for efficient similarity search. This is crucial for tasks like retrieving relevant information based on user queries, as it enables us to quickly find the most similar chunks of text in our dataset.
from langchain_community.vectorstores import FAISS
vector_db = FAISS.from_documents(texts, embeddings)

#save locally
vector_db.save_local("faiss_index")

#create a retriever to fetch relevant chunks based on user query
retriever = vector_db.as_retriever(search_kwargs={"k": 3})

#TEST
query = "What is CNN?"
relevant_chunks = retriever.invoke(query)
print(f"Number of relevant chunks: {len(relevant_chunks)}") 

for i, doc in enumerate(relevant_chunks):
    print(f"\n--- Result {i+1} ---")
    print(doc.page_content[:300])  # Print the first 300 characters of each relevant chunk
```

```python
from dotenv import load_dotenv
import os

load_dotenv()

api_key = os.getenv("OPENAI_API_KEY")
```

```python
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(
    model="gpt-3.5-turbo",
    api_key=api_key,
    temperature=0.7
)
```

```python
import os
print(os.getenv("OPENAI_API_KEY"))
```

```python
#build answer function
def ask_pdf(query):
    relevant_chunks = retriever.invoke(query)

    #create context by concatenating the relevant chunks
    context = "\n\n".join([chunk.page_content for chunk in relevant_chunks])

    #prompt
    prompt =f"""
    Answer ONLY using the context below.
    If the answer is not present, say:
    "I could not find this in the document."
    
    Context: {context}
    Question: {query}
    """
    response = llm.invoke(prompt)
    return response.content
```

```python
#test the function
answer = ask_pdf("What is CNN?")
print(answer)
```
